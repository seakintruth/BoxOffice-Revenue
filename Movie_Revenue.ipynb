```python
import numpy as np
import pandas as pd

dfOrig = pd.read_csv('movies_metadata.csv')
print("original data:")
print(dfOrig.dtypes)
print(dfOrig.shape)
```

```python
df = dfOrig[['revenue','release_date','runtime','vote_average','vote_count','popularity']]
print("selected data:")
print(df.dtypes)
```

```python
print("original shape:")
print(df.shape)

df = df.dropna()
print("after dropping NaN values:")
print(df.shape)

df = df[df.revenue > 1000000]
print("after dropping low revenue values:")
print(df.shape)
```

```python
print("originial time format:")
print(type(df['release_date'][0]))
#dfRevTime['release_date'] = dfRevTime['release_date'].toDateTime

df['release_date'] = pd.to_datetime(df['release_date'],infer_datetime_format=True)
print("fixed time format:")
print(type(df['release_date'][0]))
```
